**Ejercicio 1 :  Cargue la base de datos MNIST con target codificado ”one hot”, considerando la parte de entrenamiento, la de validación y la de test.** 

```python
# importo librerias necesarias     
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
from keras.layers import Input, Dense
from keras.models import Model
from keras.callbacks import EarlyStopping, ModelCheckpoint, Callback
```

```python
from tensorflow.examples.tutorials.mnist import input_data

# importo y guardo MNIST data
mnist = input_data.read_data_sets("./mnist/data/", one_hot=True)
```

```python
# dimensiones de la data
print(f"Imagenes de Entrenamiento: {mnist.train.images.shape}") #55000 imágenes con 784 píxeles concatenados cada una
print(f"Labels de Entrenamiento: {mnist.train.labels.shape}") #10 son la cantidad de dígitos a clasificar

print(f"Imagenes de Testeo: {mnist.test.images.shape}")
print(f"Labels de Testeo: {mnist.test.labels.shape}")
```

```python
# visualización de las imágenes
fig, ax = plt.subplots(3, 3, figsize=[15, 40])

random_image = np.random.randint(low=0, high=55000, size=1)

for j in range(3):
    for i in range(3):
        
       random_image = np.random.randint(low=0, high=55000, size=1)

       ax[i,j].imshow(mnist.train.images[random_image].reshape(28,28), 'gray') #Cambia de 784 pixeles concatenados a una matriz de 28x28
       ax[i,j].set_xticks([])
       ax[i,j].set_yticks([])
       ax[i,j].set_title('Label:  ' + str(mnist.train.labels[random_image]), color='r')


#fig.subplots_adjust(wspace=0.1, hspace=0)
plt.show()
```

**Ejercicio2 :  Defina dos  redes totalmente conectadas diferentes que sean capaces de clasificar la base MNIST. Una con 3 capas ocultas y la otra con 5. Evalúe posible overfitting en ambas y compárelas usándola parte de entrenamiento y la parte de la base destinada para validación.**

```python
## Definición de capa de entrada, capas ocultas y capa de salida del modelo

entrada = Input(shape=(784,))
layer1 = Dense(10, activation='sigmoid')(entrada)
layer2 = Dense(10, activation='sigmoid')(layer1)
layer3 = Dense(10, activation='sigmoid')(layer2)
predicciones = Dense(10, activation='softmax')(layer3)

###
```

```python
### Entrada y salida del modelo

model = Model(inputs=entrada, outputs=predicciones)
```

```python
# Definición de optimizador y función de costo
model.compile(optimizer='rmsprop', #El optimizador se aplica para ajustar los parámetros en función de la función loss
              loss='categorical_crossentropy', #Indica cuán lejos está el cálculo, luego de aplicar los parámetros, con el que debería ser la salida
              metrics=['accuracy']) #El accuracy es la fracción de imágenes que son correctamente clasificadas
```

```python
# Guardamos el modelo

filepath = "prueba1.h5"
model_checkpoint = ModelCheckpoint(filepath, monitor='loss',verbose=1, save_best_only=True)
```

```python
# Entrenamiento del modelo
history = model.fit(mnist.train.images, mnist.train.labels, # Se define el conjunto de datos de entrenamiento en forma de matrices
                    #batch size es el numero de datos que se utilizaran en cada actualizacion de los parametros, y epochs son las iteraciones 
              batch_size=32,
              epochs=50, validation_data=(mnist.test.images, mnist.test.labels), callbacks=[model_checkpoint])
```

```python
model.summary()
```

```python
## visuzalización de las curvas de entrenamiento

plt.figure()
plt.title("Accuracy vs epochs")
plt.plot( history.history['acc'] )
plt.plot( history.history['val_acc'] )
plt.legend(['Training Accuracy', 'Validation Accuracy'])


plt.figure()
plt.title("Loss vs epochs")
plt.plot( history.history['loss'] )
plt.plot( history.history['val_loss'] )
plt.legend(['Training Loss', 'Validation Loss'])

# Se puede observar que no hay overfitting, pues en la gráfica de "loss", la pérdida que se produce utilizando los datos de validación desciende hacia una línea
# estable y guarda una mínima discrepancia con la función de costo de los datos de entrenamiento. Asimismo, la gráfica de pérdida utilizando 
# los datos de entrenamiento desciende también hacia una línea estable. 
# 
```

```python
# Proceso de predicción , utilizando imágenes de testeo
predicciones = model.predict(mnist.test.images)
print(predicciones.shape)
##
```

```python
fig, ax = plt.subplots(3, 3, figsize=[15, 40])


for j in range(3):
    for i in range(3):
        
       random_image = np.random.randint(low=0, high=1000, size=1)

       ax[i,j].imshow(mnist.test.images[random_image].reshape(28,28), 'gray')
       ax[i,j].set_xticks([])
       ax[i,j].set_yticks([])
       ax[i,j].set_title('Predccion de mi red:  ' + str(np.argmax(predicciones[random_image])), color='r') # calcula el máximo valor de probabilidad 
                                                                                                           # entre los 10 dígitos

plt.show()
```

```python
# Evaluación de nuestro modelo con el conjunto de datos de testeo

test_loss, test_acc = model.evaluate(mnist.test.images, mnist.test.labels)
```

```python
print('Test accuracy:', test_acc)
```

```python
# Acertó en el 93% de las predicciones utilizando imágenes de testeo
```

***Definición de una red neuronal con 5 capas ***

```python
## Definición de capa de entrada, capas ocultas y capa de salida del modelo de
## la segunda red neuronal

Entradared2 = Input(shape=(784,))
capa1 = Dense(10, activation='sigmoid')(Entradared2)
capa2 = Dense(20, activation='sigmoid')(capa1)
capa3 = Dense(10, activation='sigmoid')(capa2)
capa4 = Dense(20, activation='sigmoid')(capa3)
capa5 = Dense(10, activation='sigmoid')(capa4)
predicciones2 = Dense(10, activation='softmax')(capa5)

###
```

```python

### Entrada y salida del modelo
model2 = Model(inputs= Entradared2, outputs=predicciones2)
```

```python
### Definición del optimizador y función de costo

model2.compile(optimizer='rmsprop', 
              loss='categorical_crossentropy', 
              metrics=['accuracy']) 
```

```python

## Guardo el modelo
filepath2 = "prueba2.h5"
model_checkpoint2 = ModelCheckpoint(filepath2, monitor='loss',verbose=1, save_best_only=True)
```

```python
# Entrenamiento del modelo
history2 = model2.fit(mnist.train.images, mnist.train.labels, 
              batch_size=32,
              epochs=50, validation_data=(mnist.test.images, mnist.test.labels), callbacks=[model_checkpoint2])
```

```python
model.summary()
```

```python
## visuzalización de las curvas de entrenamiento

plt.figure()
plt.title("Accuracy vs epochs")
plt.plot( history2.history['acc'] )
plt.plot( history2.history['val_acc'] )
plt.legend(['Training Accuracy', 'Validation Accuracy'])


plt.figure()
plt.title("Loss vs epochs")
plt.plot( history2.history['loss'] )
plt.plot( history2.history['val_loss'] )
plt.legend(['Training Loss', 'Validation Loss'])

# Se puede observar que no hay overfitting, pues en la gráfica de "loss", la pérdida que se produce utilizando los datos de validación desciende hacia una línea
# estable y guarda una mínima discrepancia con la función de costo de los datos de entrenamiento. Asimismo, la gráfica de pérdida de utilizando 
# los datos de entrenamiento desciende también hacia una línea estable. 
```

```python
# Proceso de predicción , utilizando imágenes de testeo
predicciones2 = model2.predict(mnist.test.images)
print(predicciones2.shape)
##
```
